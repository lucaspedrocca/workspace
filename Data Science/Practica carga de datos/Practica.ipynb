Importando librerias necesarias

```python
import pandas as pd
import mysql.connector
pd.options.display.max_rows = 15
```

Conectando con la base de datos

```python
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='10Caracteres.',
    database='actualizacionesWindows'
)
cursor = conn.cursor()
```

Definiendo estructura de la tabla

```python
create_table_query = '''
CREATE TABLE IF NOT EXISTS ActualizacionesJulio (
    ID INT AUTO_INCREMENT PRIMARY KEY,
    NOMBRE VARCHAR(255),
    PROPIETARIO VARCHAR(255),
    ROL VARCHAR(255),
    FECHA DATETIME,
    HORARIO_INICIO TIME,
    HORARIO_FIN TIME,
    PREBOOT VARCHAR(255),
    STATUS_PREEBOOT VARCHAR(255),
    POSTBOOT VARCHAR(255),
    STATUS_POSTBOOT VARCHAR(255),
    OBSERVACIONES VARCHAR(255),
    TICKET INT,
    EJECUTOR VARCHAR(255),
    STATUS_FINAL VARCHAR(255),
    TICKET_ANTERIOR INT    
)'''

cursor.execute(create_table_query)
conn.commit()
```

Crear tabla

```python
dataframe = pd.read_excel("ActualizacionesWindows.xlsx", header=1)

print(dataframe.columns)

dataframe = dataframe.drop(columns=['Unnamed: 15', 'Unnamed: 16','Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24','Unnamed: 25', 'Unnamed: 26', 'OK'])

dataframe.columns()

#dataframe.columns("Nombre", "Propietario", "Rol", "Fecha", "HORARIO_INICIO", "HORARIO_FIN","PREBOOT", "STATUS_PREEBOOT", "POSTBOOT", "STATUS_POSTBOOT", "OBSERVACIONES", "TICKET", "EJECUTOR", "STATUS_FINAL","TICKET_ANTERIOR" )
```

Insertar datos

```python
insert_query = """
INSERT INTO actualizacionesjulio (
    Nombre, 
    Propietario, 
    ROL, 
    Fecha, 
    Horario_de_Inicio, 
    Horario_de_Fin, 
    Preboot, 
    Status_Preboot, 
    PostBoot, 
    Status_PostBoot, 
    Observaciones, 
    Ticket, 
    Ejecutor, 
    Status_Final, 
    Ticket_Anterior
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in dataframe.iterrows():
    values = (
        row['Nombre'], 
        row['Propietario'], 
        row['ROL'], 
        row['Fecha'], 
        row['Horario_de_Inicio'], 
        row['Horario_de_Fin'], 
        row['Preboot'], 
        row['Status_Preboot'], 
        row['PostBoot'], 
        row['Status_PostBoot'], 
        row['Observaciones'], 
        row['Ticket'], 
        row['Ejecutor'], 
        row['Status_Final'], 
        row['Ticket_Anterior']
    )
    cursor.execute(insert_query, values)


conn.commit()

cursor.close()
conn.close()
```
